```python
import tensorflow as tf
import matplotlib.pyplot as plt
import os
from IPython.display import display
from ipywidgets import IntSlider, VBox, Output

from SugarcaneLeafDisease.dataset import LabeledDataset
```

### ImageLabeling

### Loading tfrecord_dataset

```python
# data loading using LabeledDataset()
data_load = LabeledDataset()
data = data_load.parsed_dataset.take(2310)

# empty list initialization
images = []
labels = []

# appending the records on the lists
for record in data:
  images.append(record['image'].numpy())
  labels.append(record['label'].numpy())

# decoding images to be able to view it on matplotlib plot
images_decoded = [tf.io.decode_image(image).numpy() for image in images]
```

#### Dynamic View

```python
output = Output() # Create an output context to manage dynamic display

# Function to dynamically display images based on slider value
def image_slider_view(change):
    labels_inversed = {0:"healthy",1:"mosaic",2:"redrot",3:"rust",4:"yellow"}
    with output:  # Ensure the output is controlled within this context
        output.clear_output(wait=True)  # Clear the previous output

        index = change['new'] # Get the selected index from the slider
        label = labels_inversed[labels[index]] # e.g. 0 -> "healthy"

        # matplotlib settings
        fig, axes = plt.subplots(1, 1, figsize=(12, 6))
        axes.imshow(images_decoded[index])
        axes.set_title(f'{label}')
        axes.axis('off')
        plt.tight_layout()
        plt.show()

# Create an interactive slider to select the image index
image_slider = IntSlider(value=0, min=0, max=2309, step=1, description='Image')

# Attach the slider to the image display function using the 'observe' method
# This ensures that the function is called whenever the slider value changes
image_slider.observe(image_slider_view, names='value')

# Display the slider and output widget together in a vertical layout
display(VBox([image_slider, output]))
```
